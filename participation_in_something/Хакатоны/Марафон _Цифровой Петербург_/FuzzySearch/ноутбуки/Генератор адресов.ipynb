```python
import random

# Списки для генерации адресов
cities = ["Санкт-Петербург", "Питер", "СПБ", "СПб", "спб", "санкт", "санкт петербург"]
cits = ["г", "город", "гор"]

streets = ["Полоцкая", "Маяковского", "Ленина", "Пушкина", "Петроградская", "Невский", "Восстания", "Гоголя", "Революции", "Щербакова"]
sts = ["ул", "улица", "пр", "пр-кт", "проспект", "пер", "переулок", "просп", "прскт"]

house_numbers = [str(i) for i in range(1, 150)]
hs = ["д", "дом"]

corpuses = [str(i) for i in range(1, 14)]
corps = ["к", "корп", "корпус"]

liters = ["А", "Б", "В", "Г"]
lits = ["лит", "литера", "литер"]

apartments = [str(i) for i in range(1, 700)]
aps = ["кв", "квартира", "кварт"]




def cut_off_end(word, cutoff_probability=0.3):
    """Случайно обрезает конец слова с заданной вероятностью"""
    if random.random() > cutoff_probability and len(word) > 4:
        cut_index = random.randint(4, len(word) - 1)
        return word[:cut_index]
    return word



# Функция для генерации случайных ошибок в словах
def introduce_typo(word, typo_probability=0.7):
    """Вводит ошибку в слово с заданной вероятностью"""
    if len(word) == 0:
        return word  # если строка пустая, не трогаем
    if random.random() > typo_probability:
        typo_word = list(word)
        typo_index = random.randint(0, len(typo_word) - 1)
        typo_word[typo_index] = random.choice("акоиощкгижбьзтПЦИЩюбуЛТПУгекуЦШЛзтужьуцдцхщЖД")
        return ''.join(typo_word)
    return word


def missing_characters(word, missing_probability=0.3):
    """Удаляет случайный символ из слова с заданной вероятностью"""
    if random.random() > missing_probability and len(word) > 2:
        missing_index = random.randint(0, len(word) - 1)
        return word[:missing_index] + word[missing_index + 1:]
    return word

def swap_characters(word, swap_probability=0.3):
    """Меняет местами случайные символы в слове с заданной вероятностью"""
    if random.random() > swap_probability and len(word) > 2:
        swap_index1 = random.randint(0, len(word) - 1)
        swap_index2 = random.randint(0, len(word) - 1)
        word_list = list(word)
        word_list[swap_index1], word_list[swap_index2] = word_list[swap_index2], word_list[swap_index1]
        return ''.join(word_list)
    return word

def format_marker(marker, space_prob, dot_prob):
        """Добавляет пробел или точку после обозначения по вероятности"""
        if not marker:
            return marker
        if random.random() > space_prob:
            marker += " "
        return marker










def generate_fuzzy_address(
        typo_prob=0.7, missing_prob=0.3, swap_prob=0.3, cut_prob=0.3, low_prob=0.6, skr_prob=0.7,
        space_prob=0.5, dot_prob=0.3,  # <-- новые параметры
        change_city=True, change_street=True, change_house=True, change_corpus=True,
        change_liter=True, change_apartment=True):
    """Генерация случайного адреса с возможностью регулировать тип и вероятность изменений"""




    city = random.choice(cities)
    if change_city:
        city = introduce_typo(city, typo_probability=typo_prob)
        city = missing_characters(city, missing_probability=missing_prob)
        city = swap_characters(city, swap_probability=swap_prob)
        city = cut_off_end(city, cutoff_probability=cut_prob)
    cit = "" if random.random() < skr_prob else random.choice(cits)
    cit = introduce_typo(cit, typo_probability=typo_prob)
    cit = format_marker(cit, space_prob, dot_prob)

    street = random.choice(streets)
    if change_street:
        street = introduce_typo(street, typo_probability=typo_prob)
        street = missing_characters(street, missing_probability=missing_prob)
        street = swap_characters(street, swap_probability=swap_prob)
        street = cut_off_end(street, cutoff_probability=cut_prob)
        street = street.lower() if random.random() > low_prob else street
    st = "" if random.random() < skr_prob else random.choice(sts)
    st = introduce_typo(st, typo_probability=typo_prob)
    st = format_marker(st, space_prob, dot_prob)

    house = random.choice(house_numbers)
    if change_house:
        house = introduce_typo(house, typo_probability=typo_prob)
        house = missing_characters(house, missing_probability=missing_prob)
    h = "" if random.random() < skr_prob else random.choice(hs)
    h = introduce_typo(h, typo_probability=typo_prob)
    h = format_marker(h, space_prob, dot_prob)

    corpus = random.choice(corpuses)
    if change_corpus:
        corpus = introduce_typo(corpus, typo_probability=typo_prob)
    corp = "" if random.random() < skr_prob +0.3 else random.choice(corps)
    corp = introduce_typo(corp, typo_probability=typo_prob)
    corp = format_marker(corp, space_prob, dot_prob)

    liter = random.choice(liters)
    if change_liter:
        liter = introduce_typo(liter, typo_probability=typo_prob)
    lit = "" if random.random() < skr_prob +0.25 else random.choice(lits)
    lit = introduce_typo(lit, typo_probability=typo_prob)
    lit = format_marker(lit, space_prob, dot_prob)

    apartment = random.choice(apartments)
    if change_apartment:
        apartment = introduce_typo(apartment, typo_probability=typo_prob)
    ap = "" if random.random() < skr_prob+0.5 else random.choice(aps)
    ap = introduce_typo(ap, typo_probability=typo_prob)
    ap = format_marker(ap, space_prob, dot_prob)






    address_parts = []

    if random.random() > 0.5:
        address_parts.append(f"{cit}{city}")

    if "проспект" in st or "просп" in st or "прскт" in st or "пр" in st or "пр-кт" in st and random.random() < 0.7:
        address_parts.append(f"{street} {st}")
    else:
        address_parts.append(f"{st}{street}")

    if random.random() > 0.5 and change_house:
        address_parts.append(f"{h}{house}")

    if corpus and random.random() > 0.5:
        address_parts.append(f"{corp}{corpus}")

    if liter and random.random() > 0.5:
        address_parts.append(f"{lit}{liter}")

    if apartment and random.random() > 0.5:
        address_parts.append(f"{ap}{apartment}")

    random.shuffle(address_parts)

    address = " ".join(address_parts)
    return address






# Генератор нескольких адресов с возможностью настройки
def generate_multiple_fuzzy_addresses(n, **kwargs):
    return [generate_fuzzy_address(**kwargs) for _ in range(n)]
```

```python
#@title Создание датасета
import random

cities = ["питер", "СПБ", "спб", "СПб", "Санкт-Петербург", "петербург"]


streets = [
    "Полоцкая", "Маяковского", "Ленина", "Пушкина", "Петроградская",
    "Невский", "Восстания", "Гоголя", "Революции", "Щербакова",
    "Кирова", "Советская", "Центральная", "Шоссейная", "Заречная",
    "Лесная", "Солнечная", "Юбилейная", "Молодёжная", "Первомайская",
    "Октябрьская", "Мира", "Гагарина", "Комсомольская", "Садовая",
    "Фрунзе", "Чкалова", "Баумана", "Толстого", "Дзержинского",
    "Карла Маркса", "Энгельса", "Победы", "Строителей", "Красная"]

cits = ["г", "город", "гор"]

streets = ["Полоцкая", "Маяковского", "Ленина", "Пушкина", "Петроградская", "Невский", "Восстания", "Гоголя", "Революции", "Щербакова"]
sts = ["ул", "улица", "пр", "пр-кт", "проспект", "пер", "переулок", "просп", "прскт"]

house_numbers = [str(i) for i in range(1, 150)]
hs = ["д", "дом"]

corpuses = [str(i) for i in range(1, 14)]
corps = ["к.", "корп", "корпус"]

liters = ["А", "Б", "В", "Г"]
lits = ["лит", "литера", "литер"]

apartments = [str(i) for i in range(1, 700)]
aps = ["кв", "квартира", "кварт", "помещ"]




def cut_off_end(word, cutoff_probability=0.3):
    """Случайно обрезает конец слова с заданной вероятностью"""
    if random.random() > cutoff_probability and len(word) > 3:
        cut_index = random.randint(4, len(word) - 1)
        return word[:cut_index]
    return word

def cut_off_end2(word, cutoff_probability=0.3):
    """Случайно обрезает конец слова с заданной вероятностью"""
    if random.random() > cutoff_probability and len(word) > 4:
        cut_index = random.randint(2, len(word) - 1)
        return word[:cut_index]
    return word


# Функция для генерации случайных ошибок в словах
def introduce_typo(word, typo_probability=0.7):
    """Вводит ошибку в слово с заданной вероятностью"""
    if len(word) == 0:
        return word  # если строка пустая, не трогаем
    if word.isdigit():
      return word
    if random.random() > typo_probability:
        typo_word = list(word)
        typo_index = random.randint(0, len(typo_word) - 1)
        typo_word[typo_index] = random.choice("акоиокгижбьзтюбугекузтужьуцдцх")
        return ''.join(typo_word)
    return word

def introduce_typo2(word, typo_probability=0.7):
    """Вводит ошибку в слово с заданной вероятностью"""
    if len(word) < 2:
        return word  # если строка пустая, не трогаем
    if random.random() > typo_probability:
        typo_word = list(word)
        typo_index = random.randint(0, len(typo_word) - 1)
        typo_word[typo_index] = random.choice("акоиокгижбьзтюбугекузтужьуцдцх")
        return ''.join(typo_word)
    return word


def missing_characters(word, missing_probability=0.3):
    """Удаляет случайный символ из слова с заданной вероятностью"""
    if random.random() > missing_probability and len(word) > 2:
        missing_index = random.randint(0, len(word) - 1)
        return word[:missing_index] + word[missing_index + 1:]
    return word

def swap_characters(word, swap_probability=0.3):
    """Меняет местами случайные символы в слове с заданной вероятностью"""
    if random.random() > swap_probability and len(word) > 2:
        swap_index1 = random.randint(0, len(word) - 1)
        swap_index2 = random.randint(0, len(word) - 1)
        word_list = list(word)
        word_list[swap_index1], word_list[swap_index2] = word_list[swap_index2], word_list[swap_index1]
        return ''.join(word_list)
    return word

def format_marker(marker, space_prob, dot_prob):
        """Добавляет пробел или точку после обозначения по вероятности"""
        if not marker:
            return marker
        if random.random() > space_prob:
            marker += " "
        return marker










def generate_fuzzy_address2(
        typo_prob=0.7, missing_prob=0.3, swap_prob=0.3, cut_prob=0.3, low_prob=0.6, skr_prob=0.7, skr_help = 1,
        space_prob=0.5, dot_prob=0.3, typo_prob2 = 1, cut_prob2 = 1,  # <-- новые параметры
        change_city=True, change_street=True, change_house=True, change_corpus=True,
        change_liter=True, change_apartment=True, change_marker=True, markers_only = False):
    """Генерация случайного адреса с возможностью регулировать тип и вероятность изменений"""




    city0 = random.choice(cities)
    if change_city:
        city = introduce_typo(city0, typo_probability=typo_prob)
        city = missing_characters(city, missing_probability=missing_prob)
        city = swap_characters(city, swap_probability=swap_prob)
        city = cut_off_end(city, cutoff_probability=cut_prob)
    cit0 = random.choice(cits)
    if change_marker:
      cit = introduce_typo2(cit0, typo_probability=typo_prob2)
      cit = cut_off_end2(cit, cutoff_probability = cut_prob2)
    cit = format_marker(cit, space_prob, dot_prob)

    street0 = random.choice(streets)
    if change_street:
        street = introduce_typo(street0, typo_probability=typo_prob)
        street = missing_characters(street, missing_probability=missing_prob)
        street = swap_characters(street, swap_probability=swap_prob)
        street = cut_off_end(street, cutoff_probability=cut_prob)
        street = street.lower() if random.random() > low_prob else street
    st0 = random.choice(sts)
    if change_marker:
      st = introduce_typo2(st0, typo_probability=typo_prob2)
      st = cut_off_end2(st, cutoff_probability = cut_prob2)
    st = format_marker(st, space_prob, dot_prob)

    house0 = random.choice(house_numbers)
    if change_house:
        house = introduce_typo(house0, typo_probability=typo_prob)
        house = missing_characters(house, missing_probability=missing_prob)
    h0 = random.choice(hs)
    if change_marker:
      h = introduce_typo2(h0, typo_probability=typo_prob2)
      h = cut_off_end2(h, cutoff_probability = cut_prob2)
    h = format_marker(h, space_prob, dot_prob)

    corpus0 = random.choice(corpuses)
    if change_corpus:
        corpus = introduce_typo(corpus0, typo_probability=typo_prob)
    corp0 = random.choice(corps)
    if change_marker:
      corp = introduce_typo2(corp0, typo_probability=typo_prob2)
      corp = cut_off_end2(corp, cutoff_probability = cut_prob2)
    corp = format_marker(corp, space_prob, dot_prob)

    liter0 = random.choice(liters)
    if change_liter:
        liter = introduce_typo(liter0, typo_probability=typo_prob)
    lit0 =  random.choice(lits)
    if change_marker:
      lit = introduce_typo2(lit0, typo_probability=typo_prob2)
      lit = cut_off_end2(lit, cutoff_probability = cut_prob2)
    lit = format_marker(lit, space_prob, dot_prob)

    apartment0 = random.choice(apartments)
    if change_apartment:
        apartment = introduce_typo(apartment0, typo_probability=typo_prob)
    ap0 = random.choice(aps)
    if change_marker:
      ap = introduce_typo2(ap0, typo_probability=typo_prob2)
      ap = cut_off_end2(ap, cutoff_probability = cut_prob2)
    ap = format_marker(ap, space_prob, dot_prob)



    if markers_only:
      city = ""
      street = ""
      house = ""
      corpus = ""
      liter = ""
      apartment = ""




    address_parts = []
    tags_parts = []

    # город
    if random.random() > 0.5:
        tag1 = get_label1(cit0)
        if random.random() < skr_prob:
            address_parts.append([city,","])
            tags_parts.append([tag1,0])
        else:
            if tag1 == 0:
              tag2 = 0
            tag2 = 0
            address_parts.append([cit, city,","])
            tags_parts.append([tag1, tag2,0])

    # улица
    if random.random() > 0:
        tag1 = get_label1(st0)
        if random.random() < skr_prob:
            address_parts.append([street,","])
            tags_parts.append([tag1,0])
        else:
            tag2 = tag1 + 1
            if ("проспект" in st or "просп" in st or "прскт" in st or "пр" in st or "пр-кт" in st) and random.random() < 0.4:
              address_parts.append([st, street,","])
              tags_parts.append([tag1, tag2,0])
            else:
              address_parts.append([street, st,","])
              tags_parts.append([tag1, tag2,0])


    # дом
    if random.random() > 0.5 and change_house:
        tag1 = get_label1(h0)
        if random.random() < skr_prob:
            address_parts.append([house,0])
            tags_parts.append([tag1,0])
        else:
            tag2 = tag1 + 1
            address_parts.append([h, house,","])
            tags_parts.append([tag1, tag2,0])

    # корпус
    if corpus and random.random() > 0.5:
        tag1 = get_label1(corp0)
        if random.random() < (skr_prob - skr_help):
            address_parts.append([corpus,","])
            tags_parts.append([tag1,0])
        else:
            tag2 = tag1 + 1
            if corp == "к." or corp == "к. ":
              address_parts.append(["к",".", corpus,","])
              tags_parts.append([tag1,0, tag2,0])
            else:
              address_parts.append([corp, corpus,","])
              tags_parts.append([tag1, tag2,0])

    # литера
    if liter and random.random() > 0.5:
        tag1 = get_label1(lit0)
        if random.random() < skr_prob:
            address_parts.append([liter,","])
            tags_parts.append([tag1,0])
        else:
            tag2 = tag1 + 1
            address_parts.append([lit, liter,"0"])
            tags_parts.append([tag1, tag2,0])

    # квартира
    if apartment and random.random() > 0.5:
        tag1 = get_label1(ap0)
        if random.random() < skr_prob - skr_help:
            address_parts.append([apartment,","])
            tags_parts.append([tag1,0])
        else:
            tag2 = tag1 + 1
            address_parts.append([ap, apartment,","])
            tags_parts.append([tag1, tag2,0])


    #print(len(address_parts))
    #print(tags_parts)
    combined = np.array(list(zip(address_parts, tags_parts)), dtype=object)
    np.random.shuffle(combined)
    address_parts, tags_parts = combined[:,0].tolist(), combined[:,1].tolist()

    flattened_address = [part for pair in address_parts for part in pair]
    flattened_tags = [part for pair in tags_parts for part in pair]


    #address = " ".join(address_parts)
    return flattened_address, flattened_tags






# Генератор нескольких адресов с возможностью настройки
def generate_multiple_fuzzy_addresses(n, **kwargs):
    return [generate_fuzzy_address2(**kwargs) for _ in range(n)]




def generate_multiple_fuzzy_addresses2(n, **kwargs):
    words_list = []
    tags_list = []

    for i in range(n):
        words, tags = generate_fuzzy_address2(**kwargs)
        words_list.append(words)
        tags_list.append(tags)

    # Формируем датафреймы
    df_words = pd.DataFrame(words_list)
    df_tags = pd.DataFrame(tags_list)

    return df_words, df_tags
```
